#  DATA EXPLORATION


```python
sns.regplot(x = df1['N'], y =df1['pH'], data = df1)
plt.show()
sns.regplot(x = df1['P'], y =df1['pH'], data = df1)
plt.show()
sns.regplot(x = df1['Ca'], y =df1['pH'], data = df1)
plt.show()
sns.regplot(x = df1['K'], y =df1['pH'], data = df1)
plt.show()
sns.regplot(x = df1['Mg'], y =df1['pH'], data = df1)
plt.show()
sns.regplot(x = df1['Na'], y =df1['pH'], data = df1)
plt.show()
```

```python
fig=plt.figure()
X=df1['N']
y=df1['pH']
ax=fig.add_axes([0,0,1,1])
ax.scatter(X,y,color='r')
ax.set_xlabel('Grades Range')
ax.set_ylabel('Grades Scored')
ax.set_title('scatter plot')
plt.show()
```

```python
fig,a =  plt.subplots(2,3)
a[0][0].bar(df1['N'],df1['pH'])
a[0][0].set_title('n vs ph')
a[0][1].bar(df1['P'],df1['pH'])
a[0][1].set_title('P vs ph')
a[0][2].bar(df1['Ca'],df1['pH'])
a[0][2].set_title('CA vs ph')

a[1][0].bar(df1['K'],df1['pH'])
a[1][0].set_title('n vs ph')
a[1][1].bar(df1['Na'],df1['pH'])
a[1][1].set_title('P vs ph')
a[1][2].bar(df1['Mg'],df1['pH'])
a[1][2].set_title('CA vs ph')


plt.show()
```
